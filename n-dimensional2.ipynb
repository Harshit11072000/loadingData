```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
df2= pd.read_csv("height_weight.csv")
```

```python
m = df2["sex"] == 2
plt.scatter(df2.loc[m, "height"], df2.loc[m, "weight"], c="blue", s=1, label="Male")
plt.scatter(df2.loc[~m, "height"], df2.loc[~m, "weight"], c="red", s=1, label="Female")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.legend(loc=2);
```

treating points with probabaility

```python
params = ["height","weight"]
male = df2.loc[m,params].values
female = df2.loc[~m,params].values
male.shape
```

```python
from chainconsumer import ChainConsumer
c = ChainConsumer()
c.add_chain(male,parameters=params,name="male",kde=1.0,color="b")
c.add_chain(male,parameters=params,name="female",kde=1.0,color="r")
c.configure(contour_labels="confidence",usetex=False)
c.plotter.plot(figsize=2.0);
```
